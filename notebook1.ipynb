```python
import numpy as np

# 1.a) Citiți imaginile din aceste fișiere și salvați-le într-un np.array (va avea dimensiunea 9x400x600).

imageList = []

for i in range(0,9):
    image = np.load("E:\\facultate\Anul 2, sem II\IA\ML\lab1\images\car_"+str(i)+".npy")
    imageList.append(image)

imageArray = np.array(imageList)

print(imageArray)
```

```python
print(np.shape(imageArray))
```

```python
# 1.b) Calculați suma valorilor pixelilor tuturor imaginilor.

print(np.sum(imageArray))
```

```python
# 1.c) Calculați suma valorilor pixelilor pentru fiecare imagine în parte.

reshapedArray = np.reshape(imageArray, (9, 240000))

sumByImage = np.sum(reshapedArray, axis = 1)

print(sumByImage)
```

```python
print(np.sum(sumByImage) == np.sum(imageArray))
```

```python
# 1.d) Afișați indexul imaginii cu suma maximă.

print(np.argmax(sumByImage))
```

```python
# 1.e) Calculați imaginea medie și afișati-o.

# cu sklearn
from skimage import io

meanImage = np.mean(reshapedArray, axis = 0)
meanImage = np.reshape(meanImage, (400, 600))

io.imshow(meanImage.astype(np.uint8))

io.show()
```

```python
# cu matplotlib

import matplotlib.pyplot as plt

# 1.e

meanImage = np.mean(reshapedArray, axis = 0)
meanImage = np.reshape(meanImage, (400, 600))

plt.imshow(meanImage, cmap='Greys_r')
```

```python
# 1.f) Cu ajutorul funcției np.std(images_array), calculați deviația standard a imaginilor.

standardDeviation = np.std(imageArray)

print(standardDeviation)
```

```python
# 1.g) . Normalizați imaginile. (se scade imaginea medie și se împarte rezultatul la deviația standard)

normalizedList = []

for i in range(0, 9):
    normalizedImage = np.divide(np.subtract(imageArray[i], meanImage), standardDeviation)
    normalizedList.append(normalizedImage)
    
    io.imshow(normalizedImage.astype(np.uint8))
    io.show()
```

```python
print(normalizedList)
```

```python
# 1.h) Decupați fiecare imagine, afișând numai liniile cuprinse între 200 și 300, respectiv coloanele cuprinse între 280 și 400.

for i in range(0, 9):
    slice = imageArray[i][200:301, 280:401]
    
    io.imshow(slice.astype(np.uint8))
    io.show()
```
